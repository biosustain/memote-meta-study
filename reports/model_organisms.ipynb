```python
import gzip
import json
from glob import glob
from operator import itemgetter
from os.path import basename
```

```python
from csv import QUOTE_NONNUMERIC
```

```python
import pandas as pd
from cobra.io import read_sbml_model
```

## BiGG

```python
files = glob("../models/bigg/*.gz")
```

```python
models = [read_sbml_model(filename) for filename in files]
```

```python
bigg = pd.DataFrame([(basename(f).split(".")[0], m.name) for f, m in zip(files, models)], columns=["model", "strain"])
```

```python
bigg.to_csv("../data/bigg/organism.csv", index=False, quoting=QUOTE_NONNUMERIC)
```

## Minho

```python
files = glob("../models/uminho/*.gz")
```

```python
len(files)
```

```python
models = []
for filename in files:
    try:
        model = read_sbml_model(filename)
    except ValueError:
        continue
    else:
        models.append(model)
```

```python
len(models)
```

```python
pd.DataFrame([(m.id, m.name) for m in models], columns=["id", "strain"])
```
